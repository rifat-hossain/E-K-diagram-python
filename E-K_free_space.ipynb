```python
# Importing required packages

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Declaration of constants
h_bar = 1.0545718e-34  # Reduced Planck constant (J*s)
m = 9.10938356e-31  # Electron mass (kg)
```

```python
K = np.linspace(-1e-24, 1e-24, 1000)
E = (h_bar * K)**2 / (2 * m)

plt.plot(K, E)
plt.xlabel('Momentum (K) [kg×m/s]')
plt.ylabel('Energy (E) [Joules]')
plt.title('E-K Diagram of an Electron in Free Space')

plt.show()
```
